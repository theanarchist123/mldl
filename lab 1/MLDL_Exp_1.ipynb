### Introduction to Python Libraries

**NumPy** (Numerical Python):
*   **Purpose**: The fundamental package for numerical computation in Python.
*   **Key Features**: Provides powerful N-dimensional array objects (arrays of numbers or items of the same data type) and sophisticated functions for mathematical operations on these arrays.
*   **Use Cases**: Essential for scientific computing, data analysis, and machine learning, forming the basis for many other libraries.

**Pandas** (Python Data Analysis Library):
*   **Purpose**: A fast, powerful, flexible, and easy-to-use open-source data analysis and manipulation tool.
*   **Key Features**: Introduces two primary data structures: `Series` (1D labeled array) and `DataFrame` (2D labeled table-like structure). Offers robust tools for reading/writing data, cleaning, transforming, and aggregating data.
*   **Use Cases**: Data cleaning, transformation, analysis, and exploratory data analysis (EDA).

**Matplotlib**:
*   **Purpose**: A comprehensive library for creating static, animated, and interactive visualizations in Python.
*   **Key Features**: Provides a vast array of plotting functions, from simple line plots to complex 3D visualizations. Allows fine-grained control over plot elements.
*   **Use Cases**: Creating various types of plots like line plots, scatter plots, bar plots, histograms, and more to visualize data.

**Seaborn**:
*   **Purpose**: A Python data visualization library based on Matplotlib, providing a high-level interface for drawing attractive and informative statistical graphics.
*   **Key Features**: Simplifies the creation of complex statistical plots, integrates well with Pandas DataFrames, and includes built-in themes and color palettes.
*   **Use Cases**: Ideal for exploring relationships between multiple variables, visualizing distributions, and making statistical plots more aesthetically pleasing with fewer lines of code.

### NumPy Example: Creating and manipulating arrays

import numpy as np

# Create a NumPy array
array_a = np.array([1, 2, 3, 4, 5])
print("Original Array A:", array_a)

# Perform a basic operation (e.g., element-wise addition)
array_b = array_a * 2
print("Array A multiplied by 2:", array_b)

# Create a 2D array (matrix)
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("\n2D Array (Matrix):\n", matrix)

# Get the shape of the array
print("Shape of Matrix:", matrix.shape)

### Pandas Example: Creating and exploring a DataFrame

```python
import pandas as pd

# Create a DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df = pd.DataFrame(data)

print("Original DataFrame:")
display(df)

# Select a column
print("\n'Age' column:")
display(df['Age'])

# Filter rows based on a condition
print("\nPeople older than 25:")
display(df[df['Age'] > 25])
```

### Matplotlib Example: Creating a simple line plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)', color='blue', linestyle='--')
plt.title('Simple Sine Wave Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()
plt.show()
```

### Seaborn Example: Creating a statistical plot (scatterplot with regression line)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset from Seaborn
tips = sns.load_dataset('tips')

print("First 5 rows of the 'tips' dataset:")
display(tips.head())

# Create a scatter plot with a regression line
plt.figure(figsize=(8, 5))
sns.regplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs. Tip (with Regression Line)')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
